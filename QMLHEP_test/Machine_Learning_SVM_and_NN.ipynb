```python
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense


file = 'QIS_EXAM_200Events.npz'
```

```python
f = np.load(file,fix_imports=True,allow_pickle=True)
f.files
train = f['training_input']
test = f['test_input']
```

```python
data1 = train.item()['0']
data2 = train.item()['1']

data3 = test.item()['0']
data4 = test.item()['1']

a = np.zeros((50,6))
a[:,:-1] = data1

b = np.ones((50,6))
b[:,:-1] = data2

c = np.zeros((50,6))
c[:,:-1] = data3

d = np.ones((50,6))
d[:,:-1] = data4

train_data = np.vstack((a,b))
test_data = np.vstack((c,d))

x_train = train_data[:,:-1]
y_train = train_data[:,-1]
x_test = test_data[:,:-1]
y_test = test_data[:,-1]
```

```python
model = SVC()
model.fit(x_train,y_train)

predictions = model.predict(x_test)
print(roc_auc_score(y_test,predictions))
```

```python
model2 = Sequential()
model2.add(Dense(8, input_dim=5, activation='relu'))
model2.add(Dense(6, activation='relu'))
model2.add(Dense(4, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
```

```python
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model2.fit(x_train, y_train, epochs=100)
```

```python
nn_predictions = model2.predict(x_test)

print(roc_auc_score(y_test,nn_predictions))
```
